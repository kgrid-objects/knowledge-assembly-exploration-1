## (Ia) A WORKSHOP ON SHARING COMPUTABLE BIOMEDICAL KNOWLEDGE AT SCALE

PROPOSITION

When scientists create and share human knowledge in a readable form, it supports MASS ACCESS. *That's good.*

But, when human knowledge is shared in a computable form, it supports [MASS ACTION](https://pmc.ncbi.nlm.nih.gov/articles/PMC6802532/). This means that when knowledge is computable, action can be taken by a diverse group of stakeholders who can more readily make use of it in predictable and repeatable ways.

A next possible step is to scale up computable knowledge access and use across many organizations. This workshop focuses on how to do just that.

---

PURPOSE

🧠 To *stimulate thinking* about how diverse stakeholders can share and benefit from computable biomedical knowledge (CBK) at scale through widespread mass action.

---

KEY MOTIVATING QUESTION

### How can CBK be widely shared *in its computable form and formats*?

---

THE PROBLEM

### The socio-technical infrastructure for sharing CBK is **incomplete**.

---

TO ASSIST A VARIETY OF STAKEHOLDERS, NOW MAY BE THE TIME TO COMBINE:

(1) the creation of scientific evidence by scientists WITH

(2) the peer-review and dissemination capabilities of publishers WITH

(3) software engineering best practices used by programmers to create reusable code WITH

(4) the code-sharing capabilities of open-source platforms WITH

(5) the knowledge management and curation expertise of librarians and information professionals

##### EXAMPLE I. AN ILLUSTRATION OF CBK STEP 1 - A "phenotype_by_diplotype" FUNCTION THAT MAPS CYP2D6 A PERSON'S GENOTYPE TO A PHENOTYPE


<img src="assets/load.png" alt="Alt text" width="15%">

````python
gene = "CYP2D6"


def phenotype_by_diplotype(inputs):
    import csv
    import os
    output = {gene: {}}
    diplotype = inputs.get(gene)
    # Calculate using existing method
    try:

        output[gene]["diplotype"] = diplotype

        csv_path = "collection/CPIC_Phenotype_CYP2D6/cpic_phenotype_cyp2d6/CYP2D6_Diplotype_Phenotype_Table.csv"
        try:
            with open(csv_path, newline="", encoding="utf-8-sig") as csvfile:
                reader = csv.DictReader(csvfile)
                output[gene]["phenotype"] = "Unknown"
                for row in reader:
                    if row["CYP2D6 Diplotype"].strip() == diplotype:
                        index = row["Coded Diplotype/Phenotype Summary"]
                        if index is None:
                            output[gene]["phenotype"] = "Unknown"
                        else:
                            output[gene]["phenotype"] = index.replace("CYP2D6 ", "").strip().capitalize()

                        break
        except Exception as csv_error:
            output[gene]["phenotype"] = f"CSV lookup error: {csv_error}"
        return output
    except Exception as error:
        return f"Error {error}"
````

##### EXAMPLE II. AN ILLUSTRATION OF CBK STEP 2 - A RULE THAT COMPUTES CODEINE RECOMMENDATIONS

<img src="assets/load.png" alt="Alt text" width="15%">

````python
# @title
import copy
import json

drug = "Codeine"
reference = {"CYP2D6": {"field": "phenotype", "value": ""}}
keysuffix = {"CYP2D6": {"positive": "", "negative": ""}}  # adjust if needed
recommendations = {
    "cyp2d6ultrarapid": {
        "implication": "Increased formation of morphine following codeine administration, leading to higher risk of toxicity",
        "recommendation": "Avoid codeine use due to potential for toxicity. Alternatives that are not affected by this CYP2D6 phenotype include morphine and nonopioid analgesics. Tramadol and, to a lesser extent, hydrocodone and oxycodone are not good alternatives because their metabolism is affected by CYP2D6 activity",
        "classification": "Strong",
    },
    "cyp2d6normal": {
        "implication": "Normal morphine formation",
        "recommendation": "Use label-recommended age- or weight-specific dosing.",
        "classification": "Strong",
    },
    "cyp2d6intermediate": {
        "implication": "Reduced morphine formation",
        "recommendation": "Use label-recommended age- or weight-specific dosing. If no response, consider alternative analgesics such as morphine or a nonopioid. Monitor tramadol use for response.",
        "classification": "Moderate",
    },
    "cyp2d6poor": {
        "implication": "Greatly reduced morphine formation following codeine administration, leading to insufficient pain relief.",
        "recommendation": "Avoid codeine use due to lack of efficacy. Alternatives that are not affected by this CYP2D6 phenotype include morphine and nonopioid analgesics. Tramadol and, to a lesser extent, hydrocodone and oxycodone are not good alternatives because their metabolism is affected by CYP2D6 activity; these agents should be avoided",
        "classification": "Strong",
    },
}


def dosingrecommendation(inputs):
    try:
        genes = {}
        output = {}
        searchkey_ready = True
        lowercase_input = {}
        search_key = ""

        # convert input keys to lowercase
        for inputkey, value in inputs.items():
            lowercase_input[inputkey.lower()] = value

        for genekey, ref in reference.items():
            key = genekey.lower()
            if key not in lowercase_input:
                break

            genes[genekey] = {}
            genes[genekey]["diplotype"] = lowercase_input[key].get("diplotype", "")
            genes[genekey]["phenotype"] = lowercase_input[key].get("phenotype", "").lower()

            targetfield = ref["field"]
            searchkey_ready = searchkey_ready and genes[genekey].get(targetfield, "") != ""

            if targetfield == "diplotype":
                if ref["value"] in genes[genekey]["diplotype"]:
                    search_key += key + ref["value"] + keysuffix[genekey]["positive"]
                else:
                    search_key += key + ref["value"] + keysuffix[genekey]["negative"]

            if targetfield == "phenotype":
                if genes[genekey]["phenotype"] != "":
                    pheno = genes[genekey]["phenotype"].replace("metabolizer", "").replace(" ", "")
                    search_key += key + pheno

        if searchkey_ready:
            if search_key in recommendations:
                output["type"] = "CPIC Recommendation"
                output["drug"] = drug
                output["genes"] = copy.deepcopy(genes)
                rec = recommendations[search_key]
                output["recommendation"] = {
                    "implication": rec["implication"],
                    "content": rec["recommendation"],
                    "classification": rec["classification"],
                }
                return output
            else:
                return f"Incorrect/invalid input for drug {drug}"
        else:
            return "Incorrect/invalid input."

    except Exception as error:
        return f"Error: {error}"

#print(json.dumps(dosingrecommendation({'CYP2D6': {'diplotype': '*3/*3', 'phenotype': 'Poor metabolizer'}}), indent=4))
````

````python
dosingrecommendation({'CYP2D6': {'phenotype': 'Poor metabolizer'}})
````

How might the potential CBK issues above be addressed?

We believe that CBK provenance, currency, testability, and maintenance can be tracked or supported.

To help **describe CBK**, we have defined some fundamental terms:

**KNOWLEDGE** - information or understanding about a subject, obtained through experience or study

**SERVICE** - a defined function or capability provided by a software system

*In addition . . .*

For **organizing CBK**, we have architected and built two types of formal modules or containers for it. They are:

**KNOWLEDGE OBJECT** - an 'Information Artifact' carrying at least one implementation of computable knowledge

**KNOWLEDGE ASSEMBLY** - a collection of Knowledge Objects with a defined purpose

#### (6a) COMBINING KNOWLEDGE OBJECTS TO CREATE A KNOWLEDGE ASSEMBLY

<img src="assets/stakeholders.png" alt="Alt text" width="8%">

Single Knowledge Objects can be narrowly focused. Stakeholders wanting to accomplish more complex tasks may wish to combine KOs. That's what Knowledge Assemblies are for.

<img src="assets/12.png" alt="Alt text" width="40%">

A Knowledge Assembly combines the CBK and CBK Services from many Knowledge Objects and has its own Metadata and assembly-level Services capable of orchestrating multi-step computations using a variety of CBK.

#### (6b) INTRODUCING THE PHARMACOGENOMICS KNOWLEDGE ASSEMBLY

As an example, consider a complex Pharmacogenomics Knowlege Assembly that combines **40** Knowledge Objects.

Building on the work on many scientists, this example KA can answer a whole panel of related drug metabolism questions for other scientists, clinicians, and patients.

The Pharmacogenomics Knowledge Assembly has two assembly-level Services. Its assembly-level services may use some or all of the Knowledge Objects it contains to compute phenotypes, drug selection, and drug dosing recommendations.

Rich metadata provide stakeholders with the information needed to identify, organize, load, understand, and execute "models" in a systematic way. (We use the term "models" here as a generalization of CBK and other inference-making artifacts.)

**GMOV** is a prototype web app that explores the "Model Player" concept.

In GMOV, "models" are KOs and KAs and other FAIR DOs. They can be loaded, inspected, combined, and executed through a common end-user interface.

#### (8b) WORKSHOP REVIEW & WRAP-UP FOCUSED ON CBK STAKEHOLDERS

- The workshop started with some useful pharmacogenomics knowledge from scientific researchers.

- It gave examples of "CBK in the Wild" ... showing how stakeholders sometimes share CBK without rich metadata today.

- It distinguished between CBK and **Services** backed by CBK.

- The workshop showed **Tests** for CBK and CBK Services .

- It also showed **Metadata** that comes from CBK stakeholders, and introduced a formal model of a Knowledge Object (KO).

- It showed how many **Knowledge Objects** can be combined into complex **Knowledge Assemblies**.

- It demo'd a **clinical app** supported by a Knowledge Assembly.

- It also showed that once CBK is organized into FAIR Digital Objects, stakeholders can use **AI chatbots** to learn about the CBK backing a clinical app.

- Finally, using our **GMOV prototype**, the workshop showed that CBK in FAIR Digital Objects can be handled like digital music files, allowing CBK to be added into AI and other systems.
